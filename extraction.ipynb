```python
import pandas as pd
import pickle
```

```python
with open('dd.pickle','rb') as f:
    dd_links = pickle.load(f)
with open('ndtv.pickle','rb') as f:
    ndtv_links = pickle.load(f)
with open('tn.pickle','rb') as f:
    tn_links = pickle.load(f)
with open('dd_date.pickle','rb') as f:
    dd_date = pickle.load(f)
```

```python
tn_links[0]
```

```python
dd = []
ndtv= []
tn = []
date = []
def check_dd(link):
    if(link[:20] == 'http://ddnews.gov.in'):
        return True
    return False

def check_ndtv(link):
    if(link[:20] == 'https://www.ndtv.com'):
        return True
    return False
def check_tn(link):
    if(link[:28] == 'https://www.timesnownews.com'):
        return True
    return False
for i in range(len(dd_links)):
    if(check_dd(dd_links[i]) and check_ndtv(ndtv_links[i]) and check_tn(tn_links[i])):
        date.append(dd_date[i])
        dd.append(dd_links[i])
        ndtv.append(ndtv_links[i])
        tn.append(tn_links[i])
```

```python
len(tn)
```

```python
dic = {
    "category" : ['politics'],
    "title": "",
    "url": "",
    "date_publish": "",
    "description": "",
    "text": ""
}
```

```python
dd_df = pd.DataFrame.from_dict(dic)
from newsplease import NewsPlease
for i,url in enumerate(dd):
    article = NewsPlease.from_url(url)
    for field in dic:
        if(field is "category"):continue
        elif(field is "date_publish"): dic[field] = date[i]
        else :dic[field] = getattr(article,field)
    dd_df = dd_df.append(dic,ignore_index = True)
dd_df = dd_df[1:]
dd_df
```

```python
dd_df
```

```python
dd_df.to_csv('ddnews.csv')
```
