```python
from newsplease import NewsPlease
import pandas as pd
import pickle
import os
import datetime
```

```python
category = "Sports"
```

```python
with open(f'pickle_files/final_links_{category}.pickle','rb') as handle:
    final_links = pickle.load(handle)
```

```python
    dd_dict = {
    "id": None,
    "date": None,
    "url": None,
    "title": None,
    "text": None,
    "ndtvId": None,
    "timesNowId": None
    }
    n_t_dict = {
    "id": None,
    "date": None,
    "url": None,
    "title": None,
    "text": None
    }
```

```python
# news = "NDTV"
path = f"/data/{category}"
os.makedirs(path,exist_ok = True)
from tqdm import tqdm_notebook
import time
```

```python
fact_df = pd.DataFrame(columns=dd_dict.keys())
if os.path.exists(path + "/factual.csv"):
    fact_df = pd.read_csv(path + "/factual.csv")
row = 0
pbar= tqdm_notebook(final_links)
for date in pbar:
    pbar.set_description("Processing %s" % date)
    for links in final_links[date]:
        if (len(fact_df) > row) and (fact_df.loc[row,"title"] != "not found"): 
            row+=1
            continue
        try:
            article = NewsPlease.from_url(links[0])
            for field in dd_dict:
                if field == "date":
                    dd_dict[field] = date
                elif field == "id" or field == "ndtvId" or field == "timesNowId":
                    dd_dict[field] = row
                else :
                    dd_dict[field] = getattr(article,field)
        except:
            dd_dict["title"] = "not found"
        for key in dd_dict.keys():
            fact_df.loc[row,key] = dd_dict[key]
        fact_df.to_csv(path + "/factual.csv")
        row += 1
```

```python
source_df = pd.DataFrame(columns=n_t_dict.keys())
if os.path.exists(path + "/ndtv.csv"):
    source_df = pd.read_csv(path + "/ndtv.csv")
row = 0
pbar = tqdm_notebook(final_links)
for date in pbar:
    pbar.set_description("Processing %s" % date)
    for links in final_links[date]:
        if (len(source_df) > row) and (source_df.loc[row,"title"] != "not found"): 
            row+=1
            continue
        try:
            article = NewsPlease.from_url(links[1])
            for field in n_t_dict:
                if field == "date":
                    n_t_dict[field] = date
                elif field == "id":
                    n_t_dict[field] = row
                else :
                    n_t_dict[field] = getattr(article,field)
        except:
            print(f"not found: {links[1]}")
            n_t_dict["title"] = "not found"
        for key in n_t_dict.keys():
            source_df.loc[row,key] = n_t_dict[key]
        source_df.to_csv(path + "/ndtv.csv")
        row += 1
```
